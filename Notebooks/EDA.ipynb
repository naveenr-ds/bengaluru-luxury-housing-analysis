Drop The Duplicate values 


```python
df.drop_duplicates(inplace=True)
```

Clean inconsistent formats

```python
df['Ticket_Price_Cr'] = (

    df['Ticket_Price_Cr']
    .str.replace(' Cr', '', regex=False) 
    .str.replace('₹', '', regex=False)  
    .astype(float) 
)
```

Handle nulls

```python

df['Unit_Size_Sqft']=df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median())
df['Ticket_Price_Cr'] = df.groupby('Unit_Size_Sqft')['Ticket_Price_Cr'].transform(lambda x: x.fillna(x.median()))
df['Amenity_Score'] = df.groupby('Micro_Market')['Amenity_Score'].transform(lambda x: x.fillna(x.median()))
df['Buyer_Comments'] = df['Buyer_Comments'].fillna(df['Buyer_Comments'].mode()[0])
```

Normalize text fields

```python
df['Micro_Market']=df['Micro_Market'].str.title() 
df.rename(columns={'Developer_Name': 'Builder'}, inplace=True) #Column Name Change For Better Analysis 
df['Builder']=df['Builder'].str.title()
df['Configuration'] = df['Configuration'].str.upper()

```

Derive columns 

```python
df['Ticket_Price_Cr']=df['Ticket_Price_Cr']*1e7
df['Price_per_Sqft']=df['Ticket_Price_Cr'] / df['Unit_Size_Sqft'] 
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
df['Purchase_Year'] = df['Purchase_Quarter'].dt.year
df['Purchase_Quarter_Num'] = df['Purchase_Quarter'].dt.quarter 
df['Purchase_Quarter_Num'] = 'Q' + df['Purchase_Quarter_Num'].astype(str)
df['Booking_Flag'] = df['Possession_Status'].apply(
    lambda x: 1 if x in ['Launch', 'Under construction'] else 0
)
```

Rename Columns 

```python
df.rename(columns={'Ticket_Price_Cr': 'Housing_Sale_Price_Cr'}, inplace=True)
```
